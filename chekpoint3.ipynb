# Data preprocessing

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
titanic = pd.read_csv('/content/sample_data/titanic-passengers.csv', sep=';')
print(titanic.columns)
print(titanic.dtypes)
titanic.head()
```

```python
print(titanic.shape)
print(titanic.size)
print(titanic.isnull().sum().sum())
print((titanic.isnull()).mean())
print(titanic.isnull().sum())
titanic.drop(['Cabin'], axis = 1, inplace = True)
titanic['Age'].fillna(round(titanic['Age'].mean()), inplace= True)
titanic['Age'] = round(titanic['Age'])
titanic
```

```python
titanic.describe()
print(titanic[titanic['Embarked'] == 'S'].count())
titanic['Embarked'].fillna('S', inplace= True)
titanic.isnull().sum()
```

```python
def plot_correlation_map( df ):

    corr = df.corr()

    s , ax = plt.subplots( figsize =( 12 , 10 ) )

    cmap = sns.diverging_palette( 220 , 10 , as_cmap = True )

    s = sns.heatmap(

        corr, 

        cmap = cmap,

        square=True, 

        cbar_kws={ 'shrink' : .9 }, 

        ax=ax, 

        annot = True, 

        annot_kws = { 'fontsize' : 12 }

        )
plot_correlation_map(titanic)
```

# * the two features fare & patch has a good correlaion .
# * Passenger ID & Parch don't have a good correlation , so they are independant.
# * Pclass & Fare have the strongest correlation .

```python
titanic["Title"] = titanic["Name"].str.split(", ").str[1].str.split(".").str[0]
titanic
```

```python
titanic['Title'].value_counts()
```

```python
titanic.drop('Name', axis = 1, inplace = True)
titanic
```

```python
# Pour etudier la correlation entre Title et les autres caractéristiques , il faut transformer les données de Title
from sklearn.preprocessing import LabelEncoder
sexx={'male' : 1 ,'female' : 0}
titanic['Sex'].replace(sexx, inplace=True)
Embarked={'S':0 , 'C':1 , 'Q':2}
titanic['Embarked'].replace(Embarked , inplace=True)
encoder = LabelEncoder()
titanic["Title_ratio"] = encoder.fit_transform(titanic["Title"])
print(titanic['Sex'].corr(titanic['Title_ratio']))
print(titanic['Age'].corr(titanic['Title_ratio']))
print(titanic['Fare'].corr(titanic['Title_ratio']))
```

```python
Title = {"Capt":"Officer", "Col":"Officer", "Major":"Officer"," Dr":"Officer", "Rev":"Officer" , "Jonkheer":"Royalty", "Don":"Royalty", "Sir" :"Royalty", "Lady" :"Royalty" , "the Countess": "Royalty", "Dona":"Royalty", "Mme":"Miss", "Mlle":"Miss", "Miss" :"Miss", "Ms":"Mrs", "Mr" :"Mrs", "Mrs" :"Mrs" , "Master" :"Master"}
titanic['Title'].replace(Title, inplace=True)
titanic['Title'].value_counts()
```

```python
titanic["Title_ratio_2"] = encoder.fit_transform(titanic["Title"])
titanic
```

```python
print(titanic['Sex'].corr(titanic['Title_ratio_2']))
print(titanic['Age'].corr(titanic['Title_ratio_2']))
print(titanic['Fare'].corr(titanic['Title_ratio_2']))
titanic.corr()
```
